# Data Cleaning And Data Transformation
 1. removing unwanted columns
 2. removing Null values from data
 3. removing particular values from columns
 4. We create column Crash_Year from CRASH_DATE

Removing Unwanted Columns

[4]

```python
del Project103["HIT_AND_RUN_I"]
del Project103["PHOTOS_TAKEN_I"]
del Project103["STATEMENTS_TAKEN_I"]
del Project103["DOORING_I"]
del Project103["WORK_ZONE_I"]
del Project103[ "WORK_ZONE_TYPE"]
del Project103["WORKERS_PRESENT_I"]
del Project103["NUM_UNITS"]
del Project103["INTERSECTION_RELATED_I"]
del Project103["NOT_RIGHT_OF_WAY_I"]
del Project103["PRIM_CONTRIBUTORY_CAUSE"]
del Project103["SEC_CONTRIBUTORY_CAUSE"]
del Project103["STREET_NO"]
del Project103["BEAT_OF_OCCURRENCE"]
del Project103["INJURIES_FATAL"]
del Project103["INJURIES_INCAPACITATING"]
del Project103["INJURIES_REPORTED_NOT_EVIDENT"]
del Project103["INJURIES_NO_INDICATION"]
del Project103["INJURIES_UNKNOWN"]
del Project103["INJURIES_NON_INCAPACITATING"]
del Project103["DATE_POLICE_NOTIFIED"]
del Project103["Time"]
del Project103["AM-PM"]
```

```python
list(Project103)
```

Removing Null Values from data

[3]

```python
Project103 = Project103.dropna()
```

```python
Project103
```

Removing particular values from columns

[2]

```python
Project103 = Project103[(Project103.ROAD_DEFECT!= 'UNKNOWN')]
```

```python
Project103 = Project103[(Project103.REPORT_TYPE!= 'NaN')]
```

```python
Project103 = Project103[(Project103.LIGHTING_CONDITION!= 'UNKNOWN')]
```

```python
Project103 = Project103[(Project103.TRAFFICWAY_TYPE!= 'DIVIDED - W/MEDIAN (NOT RAISED)')]
```

```python
Project103 = Project103[(Project103.WEATHER_CONDITION!= 'SEVERE CROSS WIND GATE')]
```

```python
Project103 = Project103[(Project103.WEATHER_CONDITION!= 'BLOWING SAND, SOIL, DIRT')]
```

```python
Project103
```

creating Column Crash_Year from CRASH_DATE

[6]

```python
Project103['Crash_Year'] = pd.DatetimeIndex(Project103['CRASH_DATE']).year
```

```python
Project103
```

```python
Project103 = Project103[(Project103.Crash_Year!= '2013')]
```

```python
Project103 = Project103[(Project103.Crash_Year!= '2014')]
```

```python
Project103
```
